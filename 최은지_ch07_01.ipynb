# 리스트요소합계

```python
aa=[]

for i in range(4):
    aa.append(0) 
print(aa)

hap=0

for i in range(0,4):
    aa[i]=int(input(str(i+1)+"번째 숫자 :"))

for i in range(0,4):
    hap+=aa[i]
print("합계는 %d" %hap)
```

```python
list1=[]
list2=[]
value=1

for i in range(3):
    for k in range(4):
        list1.append(value)
        value+=1

    print(list1)

    list2.append(list1)
    print(list2)
    list1=[]
```

```python
list1=[]
list2=[]
value=1

for i in range(3):
    for k in range(4):
        list1.append(value)
        value+=1
    list2.append(list1)
    list1=[]

for i in range(0,3):
    for k in range(0,4):
        print("%3d" %list2[i][k],end="")
    print()
```

# Self Study

```python
aa=[]

for i in range(10):
    aa.append(0)

hap=0

i=0
while i<10 :
    aa[i]=int(input(str(i+1)+"번째 숫자:"))
    hap+=aa[i]
    i+=1
print("합계는 %d" %hap)
```

```python
list1=[]
list2=[]
value=0

for i in range(4):
    for k in range(5):
        list1.append(value)
        value+=3
    list2.append(list1)
    list1=[]

for i in range(0,4):
    for k in range(0,5):
        print("%3d" %list2[i][k],end="")
    print()
```
