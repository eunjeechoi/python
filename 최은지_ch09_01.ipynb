```python
def clac(v1,v2,op):
    result=0
    if op=="+":
        result=v1+v2
    elif op=="-":
        result=v1-v2
    elif op=="*":
        result=v1*v2
    elif op=="**":
        result=v1**v2
    elif op=="/":
        if v2==0:
            result="0으로 나눌수 없습니다."
        else:
            result=v1/v2
    else:
        result ="연산자가 잘못됬습니다"
    return result

res=0
var1,var2,oper=0,0,""
var1=int(input("첫 번째 수를 입력하세요:"))
oper=input("계산을 입력하시오(+,-,*,/):")
var2=int(input("두 번째 수를 입력하세요:"))

res = clac(var1,var2,oper)

if(str(type(res))=="<class 'str'>") :
    print(res)
else:
    print("## 계산기 : %d %s %d =%d" %(var1,oper,var2,res))
```
